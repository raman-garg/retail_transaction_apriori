```python
#importinf required libraries and the dataset
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
groceries_df = pd.read_csv('groceries_subset.csv', header = None)
groceries_df.head()
```

```python
#checking the shape of dataset
groceries_df.shape
```

```python
groceries_df.info()
```

```python
# Creating a list of transactions of items
transactions = []

# Adding all the items from each row in one list leaving 1st column which is just serial number
for i in range(0, 1000):
    transactions.append([str(groceries_df.values[i,u]) for u in range(1, 24)])
```

```python
# Calculation Min_support  = 3(3 times a day) * 7 (7 days a week) / 1000 = 0.0021
# Min_confidence = it is set lower to get more relations between products
from apyori import apriori
rules = apriori(transactions, min_support=0.0021, min_confidence=0.20, min_lift=3, min_length = 2)
```

```python
# Storing the rules in a result variable
results = list(rules)

# Checking the items that were bought together with the support
results_list = []
for i in range(0, len(results)):
    results_list.append('RULE:' + str(results[i][0]) + 'SUPPORT:' + str(results[i][1]))
```
